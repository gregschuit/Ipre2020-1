# Household consumption dataset

https://machinelearningmastery.com/how-to-load-and-explore-household-electricity-usage-data/#:~:text=The%20Household%20Power%20Consumption%20dataset%20is%20a%20multivariate%20time%20series,household%20were%20collected%20every%20minute.

```python
DATASET_PATH = "./datasets/household_power_consumption.txt"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(DATASET_PATH, sep=";")
df.head()
```

```python
all_voltages = df["Voltage"]
voltages = np.array(list(datum for i, datum in enumerate(all_voltages) if i % 10000 == 0))
len(voltages)
```

```python
plt.figure(figsize=(15, 6))

plt.plot(voltages)

plt.show()
```

```python
df_1 = df[df["Date"] == "16/12/2006"]
```

```python
plt.figure(figsize=(15, 6))

feature = "Global_active_power"
# plt.plot(list(map(lambda x: x[:3], df_1["Time"])), df_1[feature])
plt.plot(df_1["Time"], df_1[feature])

plt.show()
```

```python
!find datasets/dataverse_files/*.csv -type f > dataverse_paths.txt
```

```python
with open("dataverse_paths.txt") as f:
    paths = list(map(lambda x: x.strip(), f.readlines()))
```

```python
paths
```

```python
!rm dataverse_paths.txt
```

```python
df = pd.read_csv(paths[1])
df2 = pd.read_csv(paths[2])
df.head()
```

```python
df["energy_kWh"].describe()
```

```python
plt.figure(figsize=(15, 6))

plt.plot(df[df["date"] < "2012-06-03"]["energy_kWh"].values, label="residential 1")
plt.plot(df2[df2["date"] < "2012-06-03"]["energy_kWh"].values, label="residential 2")

plt.title("energy consumed (kWh) between 2012-06-01 and 2012-06-02")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))

plt.plot(df[df["date"] < "2012-06-03"]["energy_kWh"].values, label="residential 1")
plt.plot(df2[df2["date"] == "2015-02-21"]["energy_kWh"].values, label="residential 2")

plt.title("energy consumed (kWh) between 2012-06-01 and 2012-06-02")
plt.legend()
plt.show()
```
